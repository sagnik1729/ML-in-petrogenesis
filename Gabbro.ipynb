### Box-Cox transformation then removing the mean and scaling to unit variance of all variable

```python
def standardization(data):
    data=data.abs()
    transformed_data, _ = stats.boxcox(data+1)
    transformed_data=transformed_data[transformed_data>0]
    transformed_data -= transformed_data.mean()
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(transformed_data.reshape(-1, 1))
    return scaled_data 
    
```

```python
scaled_data=standardization(df3['Y(PPM)'] )
sns.histplot(data=scaled_data , kde=True)  # You can adjust the number of bins as needed

# Customize the labels and title
plt.xlabel('Y(PPM)')
plt.ylabel('No. of Samples')
plt.title('scaling to unit variance Y(PPM)')

# Show the plot
plt.show()
```
